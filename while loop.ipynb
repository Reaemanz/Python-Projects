While Loop

```python
n = 3

while n <= 5:
    print(n)
    n = n + 1
```

```python
n = 1

while n <= 7:
    n = n + 1
    print(n)
```

```python
num = float(input("Enter a positive number: "))

while num <= 0:
    print("That's not a positive number!")
    num = float(input("Enter a positive number: "))
```
