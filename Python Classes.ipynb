Car Class

```python
class Car:
    # Constructor: runs when you create a new Car object
    def __init__(self, brand, year):
        self.brand = brand      # attribute
        self.year = year        # attribute
        self.speed = 0          # default value

    # Method: increases the car's speed
    def accelerate(self):
        self.speed += 10
        return f"Speed increased to {self.speed} km/h"

    # Method: decreases the car's speed
    def brake(self):
        self.speed -= 10
        return f"Speed decreased to {self.speed} km/h"

    # Method: returns a nice string when printing
    def __str__(self):
        return f"{self.brand} ({self.year}) moving at {self.speed} km/h"
```

How it's used

```python
# Create a car object using the constructor
my_car = Car("Toyota", 2022)

print(my_car)   # Uses __str__
print(my_car.accelerate())  # Increases speed
print(my_car.brake())       # Decreases speed
```
