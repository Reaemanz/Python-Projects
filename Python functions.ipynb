```python
# added a docstring to the multiply function

def multiply(x, y):
    """Return the product of two numbers x and y."""
    product = x * y
    return product
```

Function to convert celcious to convert temperature

```python
#  F = C * 9/5 + 32

def celsius_to_fahrenheit(celsius):
    """Convert Celsius temperature to Fahrenheit."""
    fahrenheit = celsius * 9/5 + 32
    return fahrenheit

celsius_to_fahrenheit(25)  # Example usage
```

 Write a function called cube() with one number parameter and re
turns the value of that number raised to the third power. 

```python
def cube(number):
    """Return the cube of the given number."""
    cube = number ** 3
    return cube

cube(3)  # Example usage
```
